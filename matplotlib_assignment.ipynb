1. Setup and Styles

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
print(plt.style.available)
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

2. Basic Line Plot

```python
x = np.linspace(-2, 4, 1000)
plt.plot(x, (x ** 2) - 2 * x + 1)
```

3. Multiple Lines

```python
x = np.linspace(-4, 4, 1000)
plt.plot(x, x ** 3, color='red', label='x ** 3')
plt.plot(x, np.exp(x), color='blue', label='e ** x')
plt.plot(x, np.log(x), color='green', label = 'log(x)')
plt.legend();
```

4. Styling Lines

```python
x = np.linspace(2, 3 , 1000)
plt.plot(x, x ** 2, color='green', linestyle='solid')
plt.plot(x, x ** 3, color='c', linestyle='dashed')
plt.plot(x, x ** 4, color='0.33', linestyle='dotted')
plt.plot(x, x ** 5, color='maroon', linestyle='dashdot')
```

5. Axes Limits and Scaling

```python
x = np.linspace(-10, 10, 1000)
plt.plot(x, 1 / ((x ** 2) + 1))

plt.xlim(-11, 11)
plt.ylim(-0.1, 1.1);
```

```python
plt.plot(x, 1 / ((x ** 2) + 1))

plt.xlim(11, -11)
plt.ylim(1.1, -0.1);
```

```python
plt.plot(x, 1 / ((x ** 2) + 1))
plt.axis('tight');
```

```python
plt.plot(x, 1 / ((x ** 2) + 1))
plt.axis('equal');
```

6. Labels and Legends

```python
x = np.linspace(0,50,1000)
plt.plot(x, np.sqrt(x), label='Square Root of x')
plt.plot(x, x ** (1/3), label='Cube Root of x')
plt.title("Square Root vs Cube Root")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend();
```

7. Basic Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.plot(x, y, 'o')
```

```python
plt.scatter(x, y, marker='o')
```

8. Customizing Scatter Plots

```python
x = np.linspace(0, 4 * np.pi, 50)
y = x * np.sin(x)
plt.plot(x, y, '-s', color='blue',
         markersize=7, linewidth=3,
         markerfacecolor='red',
         markeredgecolor='blue',
         markeredgewidth=1);
```

9. Combined Plot

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 16)
y = x * np.cos(x)
plt.plot(x, y, '--^', color='pink',
         markersize=10, linewidth=5,
         markerfacecolor='brown',
         markeredgecolor='pink',
         markeredgewidth=2);
```

10. Custom Plot with Alternative Style

```python
plt.style.use('ggplot')
x = np.linspace(-1.5,1.5,1000)
plt.plot(x, np.tan(x), label='tan(x)', color='orange', linestyle='dashed')
plt.plot(x, np.arctan(x), label='arctan(x)', color='c', linestyle='dotted')
plt.title("Tangent vs Inverse Tangent")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(loc='upper left')
plt.ylim(-1.5, 1.5);
```

I chose to explore the relationship between tangent and inverse tangent in this plot. The x-limit and y-limit of -1.5 to 1.5 best displayed the opposing nature of the two functions. The ggplot style of a grayscale background could allow for lighter line colors to pop visually compared to the possibilities of a white background.

11. Style Comparison

```python
plt.style.use('default')
x = np.linspace(0, 4 * np.pi, 50)
y = x * np.sin(x)
plt.plot(x, y, '-s', color='blue',
         markersize=7, linewidth=3,
         markerfacecolor='red',
         markeredgecolor='blue',
         markeredgewidth=1);
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
x = np.linspace(0, 4 * np.pi, 50)
y = x * np.sin(x)
plt.plot(x, y, '-s', color='blue',
         markersize=7, linewidth=3,
         markerfacecolor='red',
         markeredgecolor='blue',
         markeredgewidth=1);
```

```python
plt.style.use('ggplot')
x = np.linspace(0, 4 * np.pi, 50)
y = x * np.sin(x)
plt.plot(x, y, '-s', color='blue',
         markersize=7, linewidth=3,
         markerfacecolor='red',
         markeredgecolor='blue',
         markeredgewidth=1);
```

The difference between the three styles of plots I used is subtle but important. The default appearance lacks grid lines, which makes it difficult to reference where exactly the scatter points lie on the plot. The difference between the final two plots lies solely on the background color and grid color. I believe the seaborn whitegrid is the most appropriate, as the colors of the points and connecting line are vibrant enough to see on a white background, and the dark grid lines make it easier to see the location of the scatter points.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Anscombe's Quartet:
x = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
x4 = [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]
y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
```

```python
plt.scatter(x, y1)
```

```python
m, b = np.polyfit(x, y1, 1)

regression_line = m * np.array(x) + b

plt.scatter(x, y1)
plt.plot(x, regression_line,color='red')
plt.xlabel('x')
plt.ylabel('y1')
```

```python
m, b = np.polyfit(x, y2, 1)

regression_line = m * np.array(x) + b

plt.scatter(x, y2)
plt.plot(x, regression_line,color='blue')
plt.xlabel('x')
plt.ylabel('y2')
```

```python
m, b = np.polyfit(x, y3, 1)

regression_line = m * np.array(x) + b

plt.scatter(x, y3)
plt.plot(x, regression_line,color='green')
plt.xlabel('x')
plt.ylabel('y3')
```

```python
m, b = np.polyfit(x, y4, 1)

regression_line = m * np.array(x) + b

plt.scatter(x, y4)
plt.plot(x, regression_line,color='purple')
plt.xlabel('x')
plt.ylabel('y4')
```

```python
m, b = np.polyfit(x4, y4, 1)

regression_line = m * np.array(x) + b

plt.scatter(x4, y4)
plt.plot(x, regression_line,color='orange')
plt.xlabel('x4')
plt.ylabel('y4')
```

Reflection:

Although having completely unique distribution of data points, four of the five plots have identical regression lines. It appears that similar summary statistics can be reached through quite dissimilar datasets. The outlier in the `x4, y4` dataset caused the regression line to show a positive correlation between the variables when nearly all of the `x` values were identical. The outlier in the `x, y4` dataset caused the regression line to be different than the other four and show a negative correlation between the variables. It is important to visualize data in addition to calculating summary statistics because in cases like Anscombe's Quartet, datasets that would appear identical when viewing summary statistics alone may require visualization to properly identify how they differ, especially when outliers are manipulating those summary statistics.
